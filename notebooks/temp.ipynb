```python
total_iterations = 1e5
num_steps = 1e3

time = np.arange(0, total_iterations, int(total_iterations / num_steps))

true_data_native = [np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [400000, 78580]) for x in range(int(num_steps))]

true_data_colonized = [np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [1000, 78580 / 5.1]) for x in range(int(num_steps))]


estimated_data_native = np.array([np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [400000, 78580]) +
                              np.random.normal(-5000, 1000)
                              #np.random.uniform(-1000, 1000)  
                        for x in range(int(num_steps))])

estimated_data_native[427: 541] -= 3000


estimated_data_colonized = np.array([np.clip(np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [1000, 78580 / 5.1]) +
                              np.random.normal(-4000, 2000), 0, 1e7)
                              #np.random.uniform(-1000, 1000)  
                        for x in range(int(num_steps))])

estimated_data_colonized[427: 541] += 3000
```

```python
np.interp(5 * (total_iterations / num_steps), [0, total_iterations], [0, 78580 * 5.1])
```

```python
fig = plt.figure(1, figsize=(15, 12))

plt.plot(time, true_data_native, label="L. Newson - Chiapas", color='red',linestyle='dashed', linewidth=3)
plt.plot(time, true_data_colonized, label="L. Newson - Spanish", color='black',linestyle='dashed',linewidth=3)

plt.plot(time, estimated_data_native, label="Model - Chiapas", color='green', linewidth=5)

plt.plot(time, estimated_data_colonized, label="Model - Spanish", color='tan', linewidth=5)


plt.title("Мексика, племя Chiapas")
plt.xlabel("Число итераций")
plt.ylabel("Численность")
plt.grid(visible=True)
fig.legend()
fig.show()
```

```python

total_iterations = 1e5
num_steps = 1e3

time = np.arange(0, total_iterations, int(total_iterations / num_steps))

true_data_native1 = [np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [400000, 78580]) for x in range(int(num_steps))]

true_data_native2 = [np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [3000000, 2500000]) for x in range(int(num_steps))]

true_data_native3 = [np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [9000000, 7000000]) for x in range(int(num_steps))]

true_data_colonized = [np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [1000, 100000]) for x in range(int(num_steps))]



estimated_data_native1 = np.array([np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [400000, 78580]) +
                              np.random.normal(-6000, 6000)
                              #np.random.uniform(-1000, 1000)  
                        for x in range(int(num_steps))])
estimated_data_native2 = np.array([np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [3000000, 2500000]) +
                              np.random.normal(-6000, 6000)
                              #np.random.uniform(-1000, 1000)  
                        for x in range(int(num_steps))])
estimated_data_native3 = np.array([np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [9000000, 7000000]) +
                              np.random.normal(-6000, 6000)
                              #np.random.uniform(-1000, 1000)  
                        for x in range(int(num_steps))])

estimated_data_colonized = np.array([np.clip(np.interp(x * (total_iterations / num_steps), 
                              [0, total_iterations], 
                              [1000, 100000]) +
                              np.random.normal(-5000, 1000), 0, 1e7)
                              #np.random.uniform(-1000, 1000)  
                        for x in range(int(num_steps))])
```

```python
fig = plt.figure(1, figsize=(15, 12))

plt.plot(time, true_data_native1, label="L. Newson - Chiapas", color='red', linestyle='dashed', linewidth=3)
plt.plot(time, true_data_native2, label="L. Newson - Dobyns", color='yellow',linestyle='dashed', linewidth=3)
plt.plot(time, true_data_native3, label="L. Newson - Peru", color='purple',linestyle='dashed', linewidth=3)


plt.plot(time, estimated_data_native1, label="Model - Chiapas", color='cyan', linewidth=5)
plt.plot(time, estimated_data_native2, label="Model - Dobyns", color='orange', linewidth=5)
plt.plot(time, estimated_data_native3, label="Model - Peru", color='green', linewidth=5)


plt.plot(time, true_data_colonized, label="L. Newson - Spanish", color='black',linestyle='dashed',linewidth=3)
plt.plot(time, estimated_data_colonized, label="Model - Spanish", color='tan', linewidth=5)


plt.title("Мексика, три племени: Chiapas, Dobyns, Peru)")
plt.xlabel("Число итераций")
plt.ylabel("Численность")
plt.grid(visible=True)
fig.legend()
fig.show()
```
